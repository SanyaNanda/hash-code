```python
import math
def wont_get_pizza(total_pizza,t2,t3,t4,p):
    t=total_pizza-p
    if t>=0:
        t2=t2-1
    else:
        t=abs(t)
        if t>4*t4:
            t=t-4*t4
            t4=0
            t3=math.floor((3*t3-t)/3)
        else:
            t4=math.floor((4*t4-t)/4)
    return t2,t3,t4
```
