```python
def schedule(end):
    fin=[]
    for i in range(len(end)):
        ints=[]
        ints.append(end['end'][i])
        
        for j in range(len(end['names'][i])):
            ints.append(end['names'][i][j])
            ints.append(str(1))
        fin.append(ints)
    return fin
```
