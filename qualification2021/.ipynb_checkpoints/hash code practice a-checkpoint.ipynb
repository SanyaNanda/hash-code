```python
f = open("b.txt", "r")
content = f.read()
content_list = content.splitlines()
f.close()
```

```python
t = content_list[0]
t = t.split()
print(t)
```

```python
duration=int(t[0])
intersection=int(t[1])
streets=int(t[2])
cars=int(t[3])
bonus=int(t[4])
```

```python
starting=[]
ending=[]
name=[]
L=[]
```

```python
for i in range(1,streets):
    z=content_list[i].split()
    starting.append(int(z[0]))
    ending.append(int(z[1]))
    name.append((z[2]))
    L.append(z[3])
```

```python
import pandas as pd
```

```python
df2 = pd.DataFrame(list(zip(starting,ending,name,L)),columns=['start','end','name','time'])
```

```python
df3 = df2.groupby('start')['name'].apply(list).reset_index(name='names')
df3
```

```python
df4 = df2.groupby('start')['time'].apply(list).reset_index(name='times')
df4
```

```python
df5 = df2.groupby('end')['name'].apply(list).reset_index(name='names')
df5['frequency'] = [len(x) for x in df5['names']]
df5 = df5.sort_values(by=['frequency'])
df5
```

```python
no_streets_covered=[]
path=[]
endof=[]
dic={}
car_no=[]
```

```python
j=0
for i in range(streets+1,streets+cars+1):
    z=content_list[i].split()
    no_streets_covered.append(int(z[0]))
    path.append((z[1:]))
    endof.append((z[1]))
    car_no.append(j)
    j=j+1
```

```python
dic=dict(sorted(dic.items(), key=lambda item: item[1]))
```

```python
dic
```

```python
df1 = pd.DataFrame(list(zip(no_streets_covered,endof,path,car_no)),columns=['no_streets_covered','endof','path', 'car_no'])
```

```python
df1
```

```python
df1 = df1.sort_values(by=['no_streets_covered'])
df1
```

```python
df = df1.groupby('endof')['car_no'].apply(list).reset_index(name='car nos')
df
```
